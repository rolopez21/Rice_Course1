```python
import plotly.express as px
import pandas as pd
from pathlib import Path
```

## Plot Foreclosures

```python
# Read in data
foreclosures = pd.read_csv(
    Path("../../Resources/alleghany_foreclosures.csv"),
    infer_datetime_format=True,
    parse_dates=True,
    index_col="filing_date",
)

# Slice data and group
foreclosures_grp = (
    foreclosures[["municipality", "amount"]]
    .groupby([foreclosures.index.year, "municipality"])
    .count()
    .reset_index()
)
foreclosures_grp.head()
```

```python
foreclosures[["municipality", "amount"]].groupby([foreclosures.index.year, "municipality"]).count()
```

### Plot relationship between municipality and the number of foreclosures

```python
# Create scatter plot
px.scatter(foreclosures_grp, x="municipality", y="amount", color="filing_date")
```

### Plot the relationship between filing date and municipality

```python
# Create scatter plot
px.scatter(foreclosures_grp, x="municipality", y="filing_date", size="amount")
```

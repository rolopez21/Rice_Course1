```python
# Initial imports
import os
import pandas as pd
from newsapi import NewsApiClient

# pip install newsapi-python
```

```python
# Reading the News API key enviroment variable
api_key = os.getenv("news_api")

```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch top news articles about oil
oil_headlines = newsapi.get_everything(q="oil", sources='bbc-news,the-verge',language="en")
oil_headlines
```

```python
# Print total articles
print(len(oil_headlines["articles"]))
# Show sample article
print(oil_headlines["articles"][10])
```

```python
# Transform the response dictionary to a DataFrame
oil_df = pd.DataFrame(oil_headlines["articles"])
oil_df.head()
```
